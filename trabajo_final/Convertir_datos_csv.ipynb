## DICCIONARIO DE VARIABLES

- label: nombre del plato -- ['label']
- healthLabels: etiquetas de la comida -- ['healthLabels']
- ingredientLines: ingredientes de la sopa -- ['ingredientLines']
- calories: calorias de la sopa -- ['calories']
- totalTime: tiempo de cocci칩n -- ['totalTime']
- Energy: energia de la sopa --  ['totalNutrients']['ENERC_KCAL']['quantity']
- Fat: grasa de la sopa --  ['totalNutrients']['FAT']['quantity']
- Saturated: grasas saturadas --  ['totalNutrients']['FASAT']['quantity']
- Carbs: carbohidratos --  ['totalNutrients']['CHOCDF']['quantity']
- Protein: proteina --  ['totalNutrients']['PROCNT']['quantity']

```python
lista_recetas = glob.glob("delaflor_mendoza/sopas/*.txt")

df_recetas = pd.DataFrame(columns=['label','healthLabels','ingredientLines','calories','totalTime','Energy','Fat',
                           'Saturated','Carbs','Proteina'])
i = 0
for receta in lista_recetas:
    with open(receta) as json_file:  
        data = json.load(json_file)
        lista = []
        lista.append(data['label'])
        lista.append(data['healthLabels'])
        lista.append(data['ingredientLines'])
        lista.append(data['calories'])
        lista.append(data['totalTime'])
        lista.append(data['totalNutrients']['ENERC_KCAL']['quantity'])
        lista.append(data['totalNutrients']['FAT']['quantity'])
        lista.append(data['totalNutrients']['FASAT']['quantity'])
        lista.append(data['totalNutrients']['CHOCDF']['quantity'])
        lista.append(data['totalNutrients']['PROCNT']['quantity'])
        df_recetas.loc[i] = lista
        i=i+1
df_recetas.to_csv('recetas.csv',index=False)
```

## DICCIONARIO DE VARIABLES

- Title: titulo de la pelicula -- ['Title']
- Year: a침o estreno de la pelicula -- ['Year']
- Rated: quien puede ver la pelicula -- ['Rated']

- Genre: genero de la pelicula-- ['Genre']

```python
lista_peliculas = glob.glob("randy_marcial/Proyecto/peliculas/*.txt")

df_peliculas = pd.DataFrame(columns=['Title','Year','Rated','Released','Genre','Director','Plot',
                           'imdbRating','Metascore','Country'])
i = 0
for pelicula in lista_peliculas:
    with open(pelicula) as json_file:  
        data = json.load(json_file)
        lista = []
        lista.append(data['Title'])
        lista.append(data['Year'])
        lista.append(data['Rated'])
        lista.append(data['Released'])
        lista.append(data['Genre'])
        lista.append(data['Director'])
        lista.append(data['Plot'])
        lista.append(data['imdbRating'])
        lista.append(data['Metascore'])
        lista.append(data['Country'])
        df_peliculas.loc[i] = lista
        i=i+1
df_peliculas.to_csv('peliculas.csv',index=False)
```
